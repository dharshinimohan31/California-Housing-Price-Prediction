```python
num_cols = df.select_dtypes(include=['float64','int']).columns
cat_cols = df.select_dtypes(include=['object']).columns

num_cols, cat_cols
```

```python
df = df.drop_duplicates()
```

```python
num_cols
```

```python
for col in num_cols:
    df[col].fillna(df[col].median(), inplace=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

for col in num_cols:
    plt.figure(figsize=(6,4))
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot of {col}")
    plt.show()
```

```python
def remove_outliers(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5*IQR
    upper = Q3 + 1.5*IQR
    df[col] = df[col].clip(lower, upper)  # capping
    return df
```

```python
df
```

```python
num_cols
```

```python

for col in num_cols:
    df = remove_outliers(df, col)
```

```python
num_cols = num_cols.drop('median_house_value')
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df[num_cols])
```

```python
scaleroutput = StandardScaler()
y= scaleroutput.fit_transform(df[['median_house_value']])
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(handle_unknown='ignore',sparse_output=False)

df_onehot = ohe.fit_transform(df[cat_cols])
```

```python
df_scaled
```

```python
scaled_df=pd.DataFrame(df_scaled,columns=num_cols)
```

```python
scaled_df
```

```python
encoded_df=pd.DataFrame(df_onehot,columns=ohe.get_feature_names_out())
```

```python
encoded_df
```

```python
X_df=pd.concat([scaled_df,encoded_df],axis=1)
```

```python
X_df
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_df,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import numpy as np
import pandas as pd
```

```python
r2 = r2_score(y_test, model.predict(X_test))
mse = mean_squared_error(y_test, model.predict(X_test))
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, model.predict(X_test))
```

```python
print(f"R²: {r2:.2f}")
print(f"MSE: {mse:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"MAE: {mae:.2f}")
```
